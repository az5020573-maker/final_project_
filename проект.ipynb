```python
num_cols = ['top_speed_kmh', 'battery_capacity_kWh', 'number_of_cells', 'torque_nm',
            'efficiency_wh_per_km', 'range_km', 'acceleration_0_100_s',
            'fast_charging_power_kw_dc', 'towing_capacity_kg', 'cargo_volume_l',
            'seats', 'length_mm', 'width_mm', 'height_mm']

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.describe()
```

```python
# Проверка на выбросы с IQR для range_km:
Q1 = df['range_km'].quantile(0.25)
Q3 = df['range_km'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['range_km'] < (Q1 - 1.5 * IQR)) | (df['range_km'] > (Q3 + 1.5 * IQR))]
print(f"Выбросы в range_km: {len(outliers)} записей")
```

```python
# Распределение запаса хода:
plt.figure(figsize=(10,6))
sns.histplot(df['range_km'], kde=True)
plt.title('Распределение запаса хода (range_km)')
plt.xlabel('Запас хода (км)')
plt.show()
```

```python
# Корреляция между емкостью батареи и запасом хода:
plt.figure(figsize=(8,5))
sns.scatterplot(x='battery_capacity_kWh', y='range_km', data=df)
plt.title('Запас хода vs Емкость батареи')
plt.xlabel('Емкость батареи (кВтч)')
plt.ylabel('Запас хода (км)')
plt.show()
```

```python
df['brand_group'] = df['brand'].apply(lambda x: 'Tesla' if x == 'Tesla' else 'Other')
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='battery_capacity_kWh', y='range_km', hue='brand_group', data=df, palette={'Tesla': 'red', 'Other': 'blue'})
plt.title('Tesla vs Остальные: Запас хода vs Емкость батареи')
plt.xlabel('Емкость батареи (кВтч)')
plt.ylabel('Запас хода (км)')
plt.legend(title='Бренд')
plt.show()
```

```python
# Корреляционная матрица:
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Корреляционная матрица числовых признаков')
plt.show()
```

```python
# Средний запас хода по брендам (топ 10 по числу моделей):
top_brands = df['brand'].value_counts().nlargest(10).index
plt.figure(figsize=(12,6))
sns.barplot(x='brand', y='range_km', data=df[df['brand'].isin(top_brands)])
plt.title('Средний запас хода по брендам (топ 10)')
plt.ylabel('Средний запас хода (км)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Есть ли статистически значимая разница в запасе хода между двумя брендами:
brand1 = df[df['brand'] == top_brands[0]]['range_km']
brand2 = df[df['brand'] == top_brands[1]]['range_km']

t_stat, p_val = stats.ttest_ind(brand1, brand2, equal_var=False)
print(f"T-тест между {top_brands[0]} и {top_brands[1]}: t={t_stat:.3f}, p={p_val:.3f}")
if p_val < 0.05:
    print("Разница значима")
else:
    print("Разница не значима")
```
